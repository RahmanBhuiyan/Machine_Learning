```python
import pandas as pd
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
df.head(5)
```

```python
df.describe()
```

```python
df.price.mean()
```

```python
df["zscore_price"]=(df.price - df.price.mean())/df.price.std()
```

```python
df.head(2)
```

```python
outlier=df[(df.zscore_price < -3) | (df.zscore_price > 3)]
outlier
```

```python
df_no_outlier=df[(df.zscore_price > -3) | (df.zscore_price < 3)]
df_no_outlier
```

```python
df_no_outlier.price.isnull()
```

```python
nan_value=df_no_outlier.price.fillna(df_no_outlier.price.median())
nan_value.mean()
```
