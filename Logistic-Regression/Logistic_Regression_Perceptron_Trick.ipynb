```python
 def perceptron(x,y):
  x=np.insert(x,0,1,axis=1) #add extra ona column in dataset (wO=1)
  weights=np.ones(x.shape[1])#wights (w0,w1,w2,w3-----wn) value insert 1
  lr=0.1

  for i in range(1000):
    j = np.random.randint(0,100)#randomly peck value
    y_hat = step(np.dot(x[j],weights))#multifly wights and row
    weights = weights + lr*(y[j]-y_hat)*x[j]
  return weights[0],weights[1:]
```

```python
intercept,cofe=perceptron(x,y)
```

```python
def perceptron(X,y):
  
  m = []
  b = []
  
  X = np.insert(X,0,1,axis=1)
  weights = np.ones(X.shape[1])
  lr = 0.1
  
  for i in range(200):
      j = np.random.randint(0,100)
      y_hat = step(np.dot(X[j],weights))
      weights = weights + lr*(y[j]-y_hat)*X[j]
      
      m.append(-(weights[1]/weights[2]))
      b.append(-(weights[0]/weights[2]))
      
  return m,b
```

```python
m,b = perceptron(x,y)
```
