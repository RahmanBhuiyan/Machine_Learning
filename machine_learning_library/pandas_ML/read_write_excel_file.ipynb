```python
def convert_Excel(cell):
    if cell=="not available":
        return "chill bro"
    if cell =="n.a.":
        return None
    return cell

df = pd.read_excel("weather_data.xlsx","Sheet1",converters={
    "Excel":convert_Excel,
    "tmp":convert_Excel
})
```

```python
df
```

```python
df=df.to_excel("new_weather_data.xlsx",sheet_name="stocks",index=False)
```
