```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
def numpy(company):
    list=[]
    z=len(company)
    for i in range(z):
        list.append(i)
    return np.array(list)
num=numpy(company)
num
```

```python
# num=np.arange(len(company))
```

```python
plt.xticks(num,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.bar(num-0.2,revenue, width=0.4,label="Revenue")
plt.bar(num+0.2,profit, width=0.4,label="profit")
plt.legend()
```

```python
plt.xticks(num,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.barh(num,revenue,label="Revenue")
plt.legend()
```
