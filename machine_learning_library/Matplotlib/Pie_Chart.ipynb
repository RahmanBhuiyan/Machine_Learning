```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,shadow="bool")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,shadow="bool",radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,shadow="bool",autopct="%0.0f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,shadow="bool",autopct="%0.0f%%",explode=[0,0,0.2,0.2,0.1])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,shadow="bool",autopct="%0.0f%%",explode=[0,0,0.2,0.2,0.1],startangle=60)
plt.show()
plt.savefig("piechart_imge.png",bbox_inches="tight")
```
